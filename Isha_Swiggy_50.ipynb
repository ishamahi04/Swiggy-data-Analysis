```python
import seaborn as sns

from matplotlib import pyplot as plt

plt.figure(figsize=(8,4))

sns.histplot(data=df,x='Rating')

plt.title('Distribution of Rating')
plt.xlabel('Rating')
plt.show
```

```python
import seaborn as sns

from matplotlib import pyplot as plt

plt.figure(figsize=(8,4))

sns.histplot(data=df,x='Veg',kde=True)

plt.title('Distribution of Veg')

plt.xlabel('Veg')

plt.show
```

```python
import seaborn as sns

from matplotlib import pyplot as plt

plt.figure(figsize=(8,4))

sns.kdeplot(data=df, x='Cost for two')

plt.title('Distribution of Cost for two')

plt.xlabel('Cost for two')

plt.show()
```

```python
df.head(2)
```

```python
plt.figure(figsize=(8,5))

plt.subplot(1,2,1)
df['Cost for two'].value_counts().plot(kind='pie',autopct='%1.1f%%')

plt.title('pie chart of Cost for two')

plt.subplot(1,2,2)
df['city'].value_counts().plot(kind='pie',autopct='%1.1f%%')

plt.title('pie chart of city')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10,4))

sns.boxplot(data=df,x='city')

plt.title('Box plot for city Feature')

plt.xlabel('city')

plt.show()
```

```python
plt.figure(figsize=(12,4))

sns.barplot(data=df,x='Area',y='city',errorbar=None,palette='viridis')

plt.ylim(20)

plt.title('Swiggy_50')

plt.xlabel('Area')

plt.ylabel('city')

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.violinplot(data=df,x='Veg',y='Area')

plt.title('Violin Plot for Veg across Area')

plt.xlabel('Veg')

plt.ylabel('Area')

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.barplot(data=df,x='Area',y='Veg',hue='Area',errorbar=None,palette='viridis')

plt.ylim(20)

plt.title('Average Weekly Work Hours by Area and Veg')

plt.xlabel('Area')

plt.ylabel('Veg')

plt.legend(title='Swiggy_50')

plt.xticks(rotation=45)

plt.show()
```
